```python
'''plot & show'''
# plot 함수는 diagram 상에 점(marker)들을 그리는 함수
# 기본적으로 좌표와 좌표 사이에 라인을 그려준다.
# parameter로 x좌표 집합 array, y좌표 집합 array를 받는다.
# 예를 들어, (0,0), (6,250) 이라는 두 좌표를 잇는 선을 그리고 싶다면
# 인자로 [0,6], [0,250]을 전달한다.

''' simple example'''
import matplotlib.pyplot as plt 
import numpy as np

x_points = np.array([0,6])
y_points = np.array([0,250])

plt.plot(x_points,y_points)
plt.show()

# 점들을 잇는 line 대신 marker로만 plot을 표현하고 싶다면
# 3번째 인자로 ring을 의미하는 'o'를 전달해주면 된다.
plt.plot(x_points,y_points,'o')
plt.show()
```

```python
# multiple points를 diagram상에 표현하고 싶다면 
# x_points와 y_points의 개수만 맞춰주어 좌표들을 array로 전달하면 된다.
# y_points만 인자로 전달해주면 x_points는 y_points의 length에 맞추어 default값으로 설정된다.
# ex) x_points -> 0,1,2,3 ...

x_points = np.arange(10,15)
y_points = np.arange(5)

plt.plot(x_points,y_points)
plt.show()


plt.plot(y_points)
plt.show()
```

```python
''' about markers '''
# 1. marker 인자로 'marker|line|color'형식으로 'fmt' 라는 format string 인자를 전달할 수 있다.
# marker 'o', '*' 등이 있고, line shape으로 '-'-> 선, ':' -> 점선, '--' -> 실선 등이 있고, default로는 선을 잇지 않는다.
# color는 'r','g','b','c','m','y','k','w' 등이 있고 각각
# red, green, blue, cyan, magenta, yellow, black, white를 의미한다.
# 2. ms라는 인자를 통해 markersize를 결정할 수도 있다.
# 3. mec라는 인자를 통해 marker edge color, mfc 라는 인자를 통해 marker face color를 결정할 수도 있다.
# marker color인자들의 전달값으로 약어 말고도 '#4CAF50' 와 같은 HexaDecimal notation 또는 'hotpink'와 같은 color name등을 전달해 수많은 컬러표현을 할 수있다.
y_points = np.array([1,3,8,9])

plt.plot(y_points, "*:b",ms = 15, mec = 'y', mfc = 'c')
plt.show()
```

```python
''' about line '''
# linestyle(ls) and color(c)
# linestyle 인자값으로 'dotted' 'dashed' 또는 ':' '--' 와 같은 인자를 주어 라인의 모양을 결정할 수 있다. 
# color값 역시 marker 모양 결정할 때와 같이 다양한 인자를 전달하여 line의 색을 변경할 수 있다.
y_points = [1,3,8,7]
plt.plot(y_points,ls=':',c='hotpink')
plt.show()

# linewidth(lw)
# 인자값으로 float number를 전달하여 라인의 두께를 결정할 수 있다.
plt.plot(y_points,lw=15.2)
plt.show()

# multiple lines
# plot함수를 두번 사용하여 표현하는 방법과 하나의 plot함수에 x,y 집합 배열 pair를 두번 전달하여 표현하는 방식이 있다.
x1 = np.array([0,1,2,3])
y1 = np.array([1,3,8,6])
x2 = np.array([0,1,2,3])
y2 = np.array([3,2,8,9])
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()

plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
'''labels and grids'''
# x-axis 와 y-axis에 label을 추가할 수 있다. 또한 title을 입력할 수 있다.
# label들의 fontdict라는 인자에 font, color, size 에 관한 정보를 담은 dict를 전달할 수 있다.
# title의 위치를 loc이라는 인자를 통해 정할 수 있다.
x_points = np.array([1,2,3,5,4,2])
y_points = np.array([3,9,11,3,9,1])

fonts = {'family':'serif','color':'k','size':15}
plt.xlabel("x axis",fontdict = fonts)
plt.ylabel("y axis",fontdict = fonts)
plt.title("Title",loc = 'center',fontdict=fonts)
plt.plot(x_points,y_points)


plt.grid(axis = 'y', color ='m', linestyle = '--', linewidth = 0.5)
plt.show()

# grid 함수를 통해 격자를 표현할 수 있다.
# axis 인자를 통해 축을 결정할 수 있고 default는 x,y축 모두이다.
# grid 함수에서 color, linestyle, linewidth등을 결정 할 수 있다.
```

```python
'''subplot'''
# 한 개의 figure 상에 여러 plot을 그릴 수 있게 한다. 
# 인자로 3개를 받는데 처음 두개에 plots의 행, 열, 마지막 인자로 current plot의 index를 받는다.
# 각 plot 별로 title을 달 수 있고, suptitle함수로 전체 title을 달 수 있다.

x1 = np.array([1,2,3,7])
y1 = np.arange(10,50,10)
fonts = {'family':'serif','color':'b','size':10}

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title("plot 1",loc = 'center',fontdict=fonts)

x2 = np.array([6,7,8,9])
y2 = np.array([1,2,3,4])

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("plot 2",loc= 'center',fontdict=fonts)

x3 = np.array([12,4,24,1])
y3 = np.array([10,39,29,19])

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title("plot 3",loc= 'center',fontdict=fonts)


x4 = np.array([10,20,30,40])
y4 = np.array([1,2,7,9])

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title("plot 4",loc= 'center',fontdict=fonts)

plt.suptitle("subplots")
plt.show()
```

```python
'''scatter plot'''
# scatter function 역시 plot function과 같이 동일 길이의 x_scale과 y_scale을 인자로 받는다.
# scatter를 두번 실행하면 default로 각 scatter의 색이 다르나 color 인자에 값을 주어 다른 색상을 띄게 할 수도 있다.
# c인자에 점 개수 만큼의 색깔 str array 또는 cmap에 color map 이름 인자(nipy_spectral,viridis 등) 를 주고 점 개수 사이즈의 color array를 준다면
# 점별로 색깔을 입힐 수 있다.
# s 인자에도 마찬가지로 sizes관련 array를 전달하면 점별로 사이즈를 설정할 수 있다. alpha 인자에 0~1의 값을 주어 선명도(투명도)를 결정할 수도 있다.

x1 = np.random.randint(100,size=(100))
y1 = np.random.randint(100,size=(100))

colors = np.random.randint(100,size=(100))
sizes = 10 * colors

plt.scatter(x1,y1,c= colors, s = sizes, alpha =0.5,cmap= 'nipy_spectral')
plt.colorbar()
plt.show()
```

```python
''' bar plot'''
# bar plot도 마찬가지로 x,y values를 받는데 이 때, x에 보통 category values를 전달한다.
# vertical bar를 기본적으로 제공하고 horizontal bar는 barh()를 사용한다.
# width인자로 두께를 결정하고, barh()의 경우에는 height를 인자로 받는다.
# color 인자에 값을 전달하여 bar의 색을 정할 수 있다.

x = ['A', 'B', 'C']
y =np.array([1,2,3])

plt.bar(x,y,color='Blue',width = 0.5)
plt.show()  
```

```python
'''histogram and pie chart'''
# hist 함수로 히스토그램을 생성할 수 있다.
# 인자로 array를 전달하면 array의 분포에 따른 histogram을 생성해준다.

x = np.random.normal(174,5,1000)

plt.hist(x)
plt.show()

# pie 함수를 통해 pie chart를 그릴 수 있다.
# 인자로 array를 전달하면 인자별 값의 비율이 하나의 pie를 이루게 된다. ex) x/sum(array)
# label array와 color array를 전달할 수 있다. 또한 explode 라는 array를 전달하면 
# 인자별로 center에서 떨어진 정도를 정할 수 있다.
# shadow 인자에 boolean 값을 전달하여 pie에 그림자를 전달할 수 있다. 
# startangle 값을 전달하여 반시계 방향으로 진행하는 pie chart의 시작 angle을 정할 수 있다.
# legend()함수를 사용하면 범례(labels)를 간단히 볼 수 있다. 인자로 title값을 전달할 수도 있다.

p = np.array([3,12,19,7])
colors = ['y','b','k','r']
explode = np.array([0.2,0,0,0])
labels = ['att A','att B','att C','att D']

plt.pie(p,colors=colors,labels = labels, explode= explode,shadow=True,startangle=90)
plt.legend(title = "Atts")
plt.show()  
```
